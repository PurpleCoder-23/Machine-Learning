```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x=[1,2,3,4]
y=[2,4,6,8]

#resize your graph
plt.figure(figsize=(3,2),dpi=300)





plt.plot(x,y,color='cyan',label='2x',linewidth=2,marker='.',markersize=12,markeredgecolor='blue',linestyle='--')

#lint number 2
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'b--')

plt.title('Our first graph',fontdict={'fontname':'Comic sans Ms','fontsize':20})


plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png',dpi=300)

plt.show()
```

# bar chart

```python
labels=['A','B',"C"]
values=[1,4,2]

bars=plt.bar(labels,values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
```

# Load Fifa data

```python
fifa=pd.read_csv('./downloads/fifa_data.csv')
fifa
```

```python
fifa.head(5)
```

# Histograms

```python
bins=[40,50,60,70,80,90,100]


plt.hist(fifa.Overall,bins=bins,color='#abcdef')

plt.xticks(bins)



plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player skills in FIFA 2018')

plt.show()
```

# Pie Charts

```python
fifa['Preferred Foot']
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left','Right']
colors=['red','blue']

plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight
```

```python
fifa.Weight=[x.strip('lbs') if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light=fifa.loc[fifa.Weight< 125].count()[0]
light_medium=fifa.loc[(fifa.Weight>=125) & (fifa.Weight<159)].count()[0]
medium=fifa[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy=fifa[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy=fifa[(fifa.Weight>=200)].count()[0]

Weights=[light,light_medium,medium,medium_heavy,heavy]

labels=['under 125','125-150','150-175','175-200','over 200']
explode=(.4,.1,0,0,.4)

plt.pie(Weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']

labels=['FC Barcelona','Real Madrid']

boxes=plt.boxplot([barcelona,madrid],labels=labels,patch_artist=True)

for box in boxes['boxes']:
    #set edge color
    box.set(color='red',linewidth=2)
    
    #change fill color
    
    box.set(facecolor='blue')

plt.title('professional soccer team comparison')

plt.ylabel('fifa overall rating')

plt.show()
```
